```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("HR Data.csv")
```

```python
df.columns
```

```python
#Eliminating redundant entries.
df.drop_duplicates(inplace=True)


df.reset_index(drop=True, inplace=True)


print(df)
```

```python
#sanitizing specific columns.
df['Attrition'] = df['Attrition'].apply(lambda x: 'Yes' if x.lower() == 'yes' else 'No')


print(df)
```

```python
#Eliminate the dataset's NaN values.
df = df.dropna()
print(df)
```

```python
#Giving the columns new names.
df = df.rename(columns={
    'Age': 'Age_new',
    'Attrition': 'Attrition_new',
    'BusinessTravel': 'BusinessTravel_new',
    'DailyRate': 'DailyRate_new',
    'Department': 'Department_new'
   
})

print(df.head())
```
